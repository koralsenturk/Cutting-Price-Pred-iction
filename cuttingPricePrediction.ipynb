```python
y = temizDf["Kesim .Fiyati"].values
```

```python
x = temizDf.drop("Kesim .Fiyati",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33, random_state = 10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(11,activation = "relu"))
model.add(Dense(11,activation = "relu"))
model.add(Dense(11,activation = "relu"))
model.add(Dense(11,activation = "relu"))
model.add(Dense(1))
model.compile(optimizer = "adam",loss ="mse")
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), batch_size = 20, epochs = 300)
```

```python
kayÄ±pVerisi = pd.DataFrame(model.history.history)
```

```python
kayÄ±pVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
mean_absolute_error(y_test, tahminDizisi)
```
